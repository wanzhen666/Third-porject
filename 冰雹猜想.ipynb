```python
for i in range(2,101):
    print(i,"\n---------")
    j=int(i)
    while j>1:
        if j%2!=0:
            n=int(j)
            j=int(j*3+1)
            print(j,"=",n,"*3+1")
        elif j%2==0:
            n=int(j)
            j=int(j/2)
            print(j,"=",n,"/2")
    else:
        print("---------\n")
```
